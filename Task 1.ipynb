# Cell 1: Importing Required Libraries
# In this step, we import all the essential Python libraries needed for data handling and visualization:
# pandas → for loading and analyzing the dataset
# seaborn → for visualizing data
# matplotlib → for plotting graphs
# We also enable %matplotlib inline so that all plots appear inside the notebook.
    

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Make plots appear in the notebook
%matplotlib inline
```

# Cell 2 – Load and Inspect the Dataset
# In this step, we load the Iris dataset using Seaborn's built-in function.
 # Then we inspect the dataset to understand its structure:
 # shape → shows number of rows and columns
 # columns → lists all feature names
 # head() → displays the first 5 rows of the dataset
# This helps us confirm that the dataset is loaded correctly.


```python
# Load Iris dataset
iris = sns.load_dataset('iris')

# Inspect the dataset
print("Shape of dataset:", iris.shape)
print("Column names:", iris.columns)
print("First 5 rows:\n", iris.head())
```

# Cell 3: Exploring Dataset Structure and Summary Statistics
# In this step, we explore the dataset in more detail:
# info() → shows column names, data types, and missing values
#  describe() → provides statistical summary of numerical features (mean, std, min, max, quartiles)
# This helps us understand the nature of the data before visualization or model building.


```python
# Dataset info
print("\nDataset info:")
print(iris.info())

# Descriptive statistics
print("\nDescriptive statistics:")
print(iris.describe())
```

# Cell 4: Visualizing the Relationship Between Sepal Length and Sepal Width
# In this step, we create a scatter plot to observe how sepal length and sepal width vary across the three iris species.
# Each species is shown in a different color.
# This helps us see separation between species based on these two features.



```python

# Scatter plot: Sepal length vs Sepal width
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

# Cell 5: Pairplot for Visualizing Relationships Between All Features
# In this step, we generate a pairplot to visualize the pairwise relationships between all numerical features in the Iris dataset.
# Each species is marked in a different color.
# It helps in understanding separation between species and correlation patterns.


```python
# Pairplot for all features
sns.pairplot(iris, hue='species')
plt.show()
```

# Cell 6: Visualizing Feature Distributions with Histograms
# In this step, we plot histograms for all numerical features in the dataset.
#  Helps understand the distribution (normal, skewed, etc.)
#  Useful for detecting outliers
#  Each feature has its own histogram in a grid layout


```python
# Histograms for all features
iris.hist(figsize=(10,8), bins=15, edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

# Cell 7: Boxplots for Detecting Outliers
# In this step, we use boxplots to visualize the spread of each feature and detect any potential outliers.
# Boxplots show the median, quartiles, and extreme values
# Helps understand variability in each feature

```python
# Box plots to identify outliers
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title("Boxplot for Each Feature")
plt.show()
```
